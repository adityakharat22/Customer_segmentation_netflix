```python
engagement_features = [
    'avg_watch_time_minutes',
    'watch_sessions_per_week',
    'binge_watch_sessions',
    'completion_rate'
]
```

```python
engagement_data = data[engagement_features]
```

```python
churn_features =[
    'days_since_last_login',
    'watch_sessions_per_week',
    'account_age_months'
]
```

```python
churn_data = data[churn_features]
```

```python
content_features = [
    'favorite_genre',
    'rating_given',
    'recommendation_click_rate'
]
```

```python
content_data = data[content_features]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

engagement_scaled = scaler.fit_transform(engagement_data)
```

```python
churn_scaled = scaler.fit_transform(churn_data)
```

```python
content_data = pd.get_dummies(content_data , columns =['favorite_genre'])
```

```python
content_scaled = scaler.fit_transform(content_data)
```

```python
print(engagement_scaled[:5])
```

```python
print(churn_scaled[:5])
```

```python
print(content_scaled[:5])
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt 
```

```python
wcss =[]

for i in range(1,11):
    kmeans = KMeans(n_clusters= i, random_state = 42, n_init=10)
    kmeans.fit(engagement_scaled)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11), wcss)
plt.title('Elbow method')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show
```

```python
wcss = []

for i in range (1,11):
    kmeans = KMeans(n_clusters=i , random_state = 42,n_init=10 )
    kmeans.fit(churn_scaled)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss)
plt.show()
```

```python
wcss=[]

for i in range(1,11):
    kmeans =KMeans(n_clusters= i, random_state= 42, n_init= 10)
    kmeans.fit(content_scaled)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11), wcss)
plt.show()
    
```

```python
kmeans_eng = KMeans(n_clusters= 4 , random_state = 32 , n_init=10)

eng_clusters = kmeans_eng.fit_predict(engagement_scaled)

data['engagement_cluster'] = eng_clusters
```

```python
data['engagement_cluster'].value_counts()
```

```python
kmeans_churn = KMeans(n_clusters= 4 , random_state = 32 , n_init=10)

churn_clusters = kmeans_churn.fit_predict(churn_scaled)

data['churn_cluster'] = churn_clusters
```

```python
data['churn_cluster'].value_counts()
```

```python
kmeans_content = KMeans(n_clusters=5, random_state=42, n_init=10)

content_clusters = kmeans_content.fit_predict(content_scaled)

data['content_cluster'] = content_clusters
```

```python
data['content_cluster'].value_counts()
```

```python
engagement_summary = data.groupby('engagement_cluster')[
    ['avg_watch_time_minutes',
     'watch_sessions_per_week',
     'binge_watch_sessions',
     'completion_rate']
].mean()

print(engagement_summary)
```

```python
engagement_labels = {
    0: "Consistent Viewers",
    1: "Binge Watchers",
    2: "Casual Bingers",
    3: "Low Engagement Users"
}

data['engagement_segment'] = data['engagement_cluster'].map(engagement_labels)
```

```python
churn_summary = data.groupby('churn_cluster')[
    ['days_since_last_login',
     'watch_sessions_per_week',
     'account_age_months']
].mean()

print(churn_summary)
```

```python
churn_labels = {
    0: "At-Risk Users",
    1: "Loyal Users",
    2: "Casual Active Users",
    3: "New Active Users"
}

data['churn_segment'] = data['churn_cluster'].map(churn_labels)
```

```python
content_summary = data.groupby('content_cluster')[
    ['rating_given', 'recommendation_click_rate']
].mean()

print(content_summary)
```

```python
data.groupby('content_cluster')['favorite_genre'].value_counts()
```

```python
content_labels = {
    0: "Documentary Lovers",
    1: "Entertainment Mix Viewers",
    2: "Action-Horror Fans",
    3: "Romance Lovers",
    4: "Drama Viewers"
}

data['content_segment'] = data['content_cluster'].map(content_labels)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

engagement_pca = pca.fit_transform(engagement_scaled)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

plt.scatter(
    engagement_pca[:,0],
    engagement_pca[:,1],
    c=data['engagement_cluster'],
)

plt.title("Engagement Segmentation")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.show()
```

```python
churn_pca = pca.fit_transform(churn_scaled)

plt.figure(figsize=(8,6))

plt.scatter(
    churn_pca[:,0],
    churn_pca[:,1],
    c=data['churn_cluster'],
)

plt.title("Churn Segmentation")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.show()
```

```python
content_pca = pca.fit_transform(content_scaled)

plt.figure(figsize=(8,6))

plt.scatter(
    content_pca[:,0],
    content_pca[:,1],
    c=data['content_cluster'],
)

plt.title("Content Segmentation")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.show()
```

```python
data['engagement_segment'].value_counts().plot(kind='line')
plt.title("Engagement Segment Distribution")
plt.show()
```
